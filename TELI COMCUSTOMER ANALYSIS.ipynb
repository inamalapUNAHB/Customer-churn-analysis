```python
total_count=telco.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack()*100
#plot
fig,ax=plt.subplots(figsize=(4,4))# ajusting the figures for the better visualiztion
# plot the bars
total_count.plot(kind='bar',stacked=True,ax=ax,color=['blue','orange'])
for p in ax.patches:
    widh,height=p.get_width(),p.get_height()
    x,y=p.get_xy()
    ax.text(x+widh/2,y+height/2,f'{height:1f}%',ha='center',va='center')
plt.title('Churn by senior Citizen(stacked bar chart)')
plt.xlabel('senior citizen')
plt.ylabel('percentage')
plt.xticks(rotation=0)
plt.legend(title='Churn',loc='upper right')
plt.show()
```

```python
total_count=telco.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack()*100
#plot
fig,ax=plt.subplots(figsize=(4,4))# ajusting the figures for the better visualiztion
# plot the bars
total_count.plot(kind='bar',stacked=True,ax=ax,color=['blue','orange'])
for p in ax.patches:
    widh,height=p.get_width(),p.get_height()
    x,y=p.get_xy()
    ax.text(x+widh/2,y+height/2,f'{height:1f}%',ha='center',va='center')
plt.title('Churn by senior Citizen(stacked bar chart)')
plt.xlabel('senior citizen')
plt.ylabel('percentage')
plt.xticks(rotation=0)
plt.legend(title='Churn',bbox_to_anchor=(0.9,0.9)) # out side bbox
plt.show()
```

```python
#  'PhoneService', 'MultipleLines', 'InternetService',
     #  'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    #   'TechSupport', 'StreamingTV', 'StreamingMovies'

# List of columns to create count plots for
columns = ['PhoneService', 'MultipleLines', 'InternetService', 
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up the figure and axes for subplots
num_columns = len(columns)
fig, axes = plt.subplots(nrows=(num_columns + 2) // 3, ncols=3, figsize=(15, 10))
axes = axes.flatten()  # Flatten the 2D array of axes for easy indexing

# Create count plots for each column
for i, column in enumerate(columns):
    sns.countplot(data=telco, x=column, ax=axes[i])
    axes[i].set_title(column)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')

# Hide any unused axes
for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.show()
```
